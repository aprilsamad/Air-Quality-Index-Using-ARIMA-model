```python
df['date_time'] = pd.to_datetime(df['date_time'], format='%d-%m-%Y %H:%M')
# df['date_time'] = pd.to_datetime(df['date_time'])
df.set_index('date_time', inplace=True)

# Convert the index to DatetimeIndex
df.index = pd.DatetimeIndex(df.index)

# Extract the data for each day between 13:45 to 14:45
data_per_day = df.between_time('13:45', '14:45')
```

```python
# print(df["PM2.5_nearest"][0])
AQI_category = {'Good' : 0 ,'Moderate' : 0 , 'Unhealthy for Sensitive Groups': 0 ,'Unhealthy':0,'Very Unhealthy':0, 'Hazardous':0 }
# AQI_category['Good'] = 0
# AQI_category['Moderate'] = 0
# AQI_category['Unhealthy for Sensitive Groups'] = 0
# AQI_category['Unhealthy'] = 0
# AQI_category['Very Unhealthy'] = 0
# AQI_category['Hazardous'] = 0

for i in df["PM2.5"] :
    
    if i<=12 :
        AQI_category['Good'] += 1
    if 12< i <= 35.4:
        AQI_category['Moderate'] += 1
    if 35.4< i <= 55.4:
        AQI_category[ 'Unhealthy for Sensitive Groups'] +=1
    if 55.4< i <= 150.4:
        AQI_category['Unhealthy']+= 1
    if 150.4< i <= 250.4:
        AQI_category['Very Unhealthy']+= 1
    if  250.4 < i <= 350.4:
        AQI_category['Hazardous']+= 1
    
    
    
print(AQI_category)
```

```python
# Extract the keys and values from the dictionary
keys = list(AQI_category.keys())
values = list(AQI_category.values())

# Create the bar plot
fig  = plt.figure(figsize = (10,5))
plt.bar(keys, values)
plt.xlabel('AQI Categories')
plt.ylabel('Total_Values')
plt.title('PM2.5 air quality')
plt.xticks(rotation=45)
plt.show()

```

```python
plt.pie(values, labels=keys, autopct='%1.1f%%')

# Add a title
plt.title('Pie Chart of Dictionary Values')

# Display the chart
plt.show()
```

```python
series = df['PM2.5']

# Calculate the moving average with a window size of 3
moving_average = series.rolling(window=97).mean()

# Print the moving average
print(moving_average)
```

```python
plt.plot(series, label='Original')
plt.plot(moving_average, label='Moving Average')

# Set labels and title for the plot
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Moving Average')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
